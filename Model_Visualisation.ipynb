```python
colorMap = defaultdict(dict)
colorMap[layers.InputLayer]['fill'] = 'grey'
colorMap[layers.Conv2D]['fill'] = 'lightgrey'
colorMap[layers.Activation]['fill'] = 'white'
colorMap[layers.GlobalAveragePooling2D]['fill'] = 'black'
colorMap[layers.MaxPooling2D]['fill'] = 'pink'
colorMap[layers.UpSampling2D]['fill'] = 'purple'


visualkeras.layered_view(model, scale_xy=0.4, scale_z=100, max_z=10,legend=True, color_map=colorMap,spacing=30)
```
